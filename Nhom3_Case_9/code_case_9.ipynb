1. Đọc dữ liệu từ hai file semester1_scores.csv và semester2_scores.csv.

```python
import numpy as np

data1 = np.genfromtxt(r"D:\\18A1PNC\\semester1_scores.csv", delimiter=",", skip_header=1, dtype=float)
data2 = np.genfromtxt(r"D:\\18A1PNC\\semester2_scores.csv", delimiter=",", skip_header=1, dtype=float)
```

2. Tách student_id, midterm, final, attendance/hw thành các mảng riêng.

```python
student_id_1 = data1[:, 0]
midterm_1    = data1[:, 1]
final_1      = data1[:, 2]
att_1        = data1[:, 3]

print("ID Sinh viên (HK1):", student_id_1)
print("Điểm Giữa kỳ (HK1):", midterm_1)
print("Điểm Cuối kỳ (HK1):", final_1)
print("Điểm Chuyên cần (HK1):", att_1)
```

```python
student_id_2 = data2[:, 0]
midterm_2    = data2[:, 1]
final_2      = data2[:, 2]
att_2        = data2[:, 3]

print("ID Sinh viên (HK2):", student_id_2)
print("Điểm Giữa kỳ (HK2):", midterm_2)
print("Điểm Cuối kỳ (HK2):", final_2)
print("Điểm Chuyên cần (HK2):", att_2)
```

3. Tính điểm tổng hợp: total = 0.3*midterm + 0.5*final + 0.2*attendance.

```python
total_1 = 0.3 * midterm_1 + 0.5 * final_1 + 0.2 * att_1
print("Điểm tổng hợp học kỳ 1 (Total 1):", total_1)
```

```python
total_2 = 0.3 * midterm_2 + 0.5 * final_2 + 0.2 * att_2
print("Điểm tổng hợp học kỳ 2 (Total 2):", total_2)
```

4. Kiểm tra dtype, shape; in 5 giá trị đầu/cuối

```python
data1.dtype
data1.shape
data1[:5]
data1[-5:]

print("HỌC KỲ 1: ")
print(f"\nKiểu dữ liệu (dtype): {data1.dtype}")
print(f"Kích thước mảng (shape): {data1.shape}")
print("5 dòng đầu: ", data1[:5])
print("5 dòng cuối: ", data1[-5:])
```

```python
data2.dtype
data2.shape
data2[:5]
data2[-5:]

print("HỌC KỲ 2: ")
print(f"\nKiểu dữ liệu (dtype): {data2.dtype}")
print(f"Kích thước mảng (shape): {data2.shape}")
print("5 dòng đầu: ", data2[:5])
print("5 dòng cuối: ", data2[-5:])
```

5. Slicing: nhóm sinh viên đầu – giữa – cuối

```python
n = len(total_1)
group1 = total_1[:n//3]
group2 = total_1[n//3:2*n//3]
group3 = total_1[2*n//3:]

print("PHÂN TÍCH SLICING HỌC KỲ 1: ")
print(f"\nTổng số sinh viên: {n}")
print(f"Nhóm 1 (Đầu): {len(group1)} sinh viên. 5 giá trị đầu: {group1[:5]}")
print(f"Nhóm 2 (Giữa): {len(group2)} sinh viên. 5 giá trị đầu: {group2[:5]}")
print(f"Nhóm 3 (Cuối): {len(group3)} sinh viên. 5 giá trị đầu: {group3[:5]}")
```

```python
m = len(total_2)
group1 = total_2[:m//3]
group2 = total_2[m//3:2*m//3]
group3 = total_2[2*m//3:]

print("PHÂN TÍCH SLICING HỌC KỲ 2: ")
print(f"\nTổng số sinh viên: {m}")
print(f"Nhóm 1 (Đầu): {len(group1)} sinh viên. 5 giá trị đầu: {group1[:5]}")
print(f"Nhóm 2 (Giữa): {len(group2)} sinh viên. 5 giá trị đầu: {group2[:5]}")
print(f"Nhóm 3 (Cuối): {len(group3)} sinh viên. 5 giá trị đầu: {group3[:5]}")
```

6. Boolean mask phân loại: xuất sắc, khá, trung bình, yếu

```python
excellent_1 = total_1 >= 8.5
good_1      = (total_1 >= 7) & (total_1 < 8.5)
average_1   = (total_1 >= 5) & (total_1 < 7)
weak_1      = total_1 < 5

print("PHÂN LOẠI HỌC LỰC HỌC KỲ 1: ")
print(f"\nXuất sắc (>= 8.5): {np.sum(excellent_1)} sinh viên")
print(f"Khá (7.0 - 8.49): {np.sum(good_1)} sinh viên")
print(f"Trung bình (5.0 - 6.99): {np.sum(average_1)} sinh viên")
print(f"Yếu (< 5.0): {np.sum(weak_1)} sinh viên")
```

```python
excellent_2 = total_2 >= 8.5
good_2 = (total_2 >= 7) & (total_2 < 8.5)
average_2 = (total_2 >= 5) & (total_2 < 7)
weak_2 = total_2 < 5

print("PHÂN LOẠI HỌC LỰC HỌC KỲ 2: ")
print(f"\nXuất sắc (>= 8.5): {np.sum(excellent_2)} sinh viên")
print(f"Khá (7.0 - 8.49): {np.sum(good_2)} sinh viên")
print(f"Trung bình (5.0 - 6.99): {np.sum(average_2)} sinh viên")
print(f"Yếu (< 5.0): {np.sum(weak_2)} sinh viên")
```

7. Broadcasting: diff_total giữa hai kỳ

```python
diff_total = total_2 - total_1

mean_change = np.mean(diff_total)
std_change = np.std(diff_total)
min_change = np.min(diff_total)
max_change = np.max(diff_total)

print("THỐNG KÊ MỨC ĐỘ THAY ĐỔI TỔNG ĐIỂM (HK2 - HK1): ")
print(f"\nĐộ lệch điểm trung bình: {mean_change:.2f}")
print(f"Độ lệch độ lệch chuẩn: {std_change:.2f}")
print(f"Độ lệch thấp nhất: {min_change:.2f}")
print(f"Độ lệch cao nhất: {max_change:.2f}")
```

```python
num_improved = np.sum(diff_total > 0)
num_declined = np.sum(diff_total < 0)
num_stable = np.sum(diff_total == 0)

print(f"Số sinh viên tiến bộ (delta > 0): {num_improved}")
print(f"Số sinh viên tụt lùi (delta < 0): {num_declined}")
print(f"Số sinh viên giữ nguyên (delta = 0): {num_stable}")
```

8. Fancy indexing: top-10 theo từng học kỳ

```python
top10_idx_1 = np.argsort(total_1)[-10:][::-1]
top10_students_1 = data1[top10_idx_1]
```

```python
top10_idx_2 = np.argsort(total_2)[-10:][::-1]
top10_students_2 = data2[top10_idx_2]
```

```python
print("TOP 10 SINH VIÊN THEO TỪNG KỲ: ")
print(f"\nTOP 10 Học kỳ 1: {top10_students_1} ")
print(f"TOP 10 Học kỳ 2: {top10_students_2} ")
```

9. Phân tích thay đổi |Δtotal| > 1.5

```python
strong_change = np.abs(diff_total) > 1.5
students_strong_change = student_id_1[strong_change]
print(f"Sinh viên Tiến bộ mạnh (delta > 1.5): {students_strong_change}")
print(f"Số lượng: {len(students_strong_change)}")
```

```python
strong_decline = diff_total < -1.5 
students_strong_decline = student_id_2[strong_decline]
print(f"Sinh viên Tụt dốc mạnh (delta < -1.5): {students_strong_decline}")
print(f"Số lượng: {len(students_strong_decline)}")
```

10. Thống kê tổng hợp theo học kỳ

```python
mean_1 = np.mean(total_1)
std_1 = np.std(total_1)
min_1 = np.min(total_1)
median_1 = np.median(total_1)

print("THỐNG KÊ TỔNG ĐIỂM HỌC KỲ 1: ")
print(f"\nMean (Điểm trung bình): {mean_1:.3f}")
print(f"Std (Độ lệch chuẩn): {std_1:.3f}")
print(f"Min (Điểm thấp nhất): {min_1:.3f}")
print(f"Median (Điểm trung vị): {median_1:.3f}")
```

```python
mean_2 = np.mean(total_2)
std_2 = np.std(total_2)
min_2 = np.min(total_2)
median_2 = np.median(total_2)

print("THỐNG KÊ TỔNG ĐIỂM HỌC KỲ 2: ")
print(f"\nMean (Điểm trung bình): {mean_2:.3f}")
print(f"Std (Độ lệch chuẩn): {std_2:.3f}")
print(f"Min (Điểm thấp nhất): {min_2:.3f}")
print(f"Median (Điểm trung vị): {median_2:.3f}")
```
